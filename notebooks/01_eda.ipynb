# 01 - Exploratory Data Analysis
Heart Disease UCI dataset. Uses the reusable `heart_mlops` package.

```
import sys, os
sys.path.insert(0, os.path.abspath(os.path.join('..', 'src')))
from heart_mlops.data_download import acquire
from heart_mlops.preprocessing import clean_data
df = clean_data(acquire())
df.head()
```

## Shape, dtypes and missing values

```
print(df.shape)
print(df.dtypes)
df.isna().sum()
```

## Target class balance

```
df['target'].value_counts(normalize=True)
```

## Generate & save all EDA figures

```
from heart_mlops.eda import run_eda
run_eda()
print('Figures saved to reports/figures/')
```

Figures produced: histograms, correlation heatmap, class balance, missing values, feature-vs-target boxplots.
